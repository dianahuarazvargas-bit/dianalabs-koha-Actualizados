# **4. Limpieza de datos**



```python
import pandas as pd

# --- URLs de los Google Sheets ---
# Sheet A: el documento con los datos a integrar
url_sheet_a = 'https://docs.google.com/spreadsheets/d/1_xtluQNjQGUs7mTXiNReB4UITdR2rFGoQaIpa6lu0ds/edit?gid=181236238#gid=181236238'
csv_export_sheet_a = url_sheet_a.replace('/edit?gid=', '/export?format=csv&gid=')

# --- 1. Cargar el Google Sheet de Actualizaciones (Sheet A) ---
try:
    # Cargar Sheet A, asegurando que 'biblionumber_999_c' sea tipo string para una fusión consistente
    df_new_updates = pd.read_csv(csv_export_sheet_a, dtype={'biblionumber_999_c': str})
    print("Contenido del Google Sheet de Actualizaciones (Sheet A):")
    display(df_new_updates.head())

    # Limpiar espacios en blanco de 'biblionumber_999_c' para la fusión
    df_new_updates['biblionumber_999_c'] = df_new_updates['biblionumber_999_c'].astype(str).str.strip()

    # Renombrar 'subbranch_952_b' a '952_b' en df_new_updates para que coincida
    if 'subbranch_952_b' in df_new_updates.columns:
        df_new_updates = df_new_updates.rename(columns={'subbranch_952_b': '952_b'})
        print("\nColumna 'subbranch_952_b' renombrada a '952_b' en df_new_updates.")

except Exception as e:
    print(f"Error al leer el Google Sheet de Actualizaciones (Sheet A): {e}")
    print("Asegúrate de que la hoja de cálculo sea pública o que tu Google Drive esté montado y tengas los permisos adecuados.")
    # Detener la ejecución si no se puede cargar el archivo vital
    raise

# --- Preparar df_normalized_global (Sheet B, el documento antiguo normalizado) ---
# df_normalized_global debería estar disponible de celdas anteriores.
# Aseguramos que 'biblionumber_999_c' sea tipo string para una fusión consistente
if 'df_normalized_global' in locals() and isinstance(df_normalized_global, pd.DataFrame):
    df_current_global = df_normalized_global.copy()
    df_current_global['biblionumber_999_c'] = df_current_global['biblionumber_999_c'].astype(str).str.strip()
else:
    print("Error: 'df_normalized_global' no se encontró o no es un DataFrame. Asegúrate de ejecutar las celdas anteriores.")
    raise ValueError("'df_normalized_global' no está disponible.")

# --- 3. Filtrar df_current_global para biblionumber_999_c que coinciden (condición 1 y 3) ---
# Obtener los biblionumber_999_c únicos del sheet de actualizaciones
biblio_nums_in_updates = df_new_updates['biblionumber_999_c'].dropna().unique()

# Filtrar df_current_global para mantener solo los biblionumber_999_c presentes en df_new_updates
df_filtered_global = df_current_global[df_current_global['biblionumber_999_c'].isin(biblio_nums_in_updates)].copy()

print(f"\nDataFrame filtrado basado en {len(biblio_nums_in_updates)} biblionumbers únicos de la hoja de actualizaciones:")
display(df_filtered_global.head())
print(f"Número de filas después del filtro: {len(df_filtered_global)}")

# --- 4. Alertar sobre biblionumber_999_c en Sheet A que no están en Sheet B ---
biblio_nums_in_filtered_global = df_filtered_global['biblionumber_999_c'].dropna().unique()
missing_in_filtered_global = set(biblio_nums_in_updates) - set(biblio_nums_in_filtered_global)

if missing_in_filtered_global:
    print("\n--- ALERTA: Biblionumbers en la hoja de actualizaciones no encontrados en el documento global ---")
    for biblio_num in sorted(list(missing_in_filtered_global)):
        print(f"- {biblio_num}")
else:
    print("\nTodos los biblionumbers de la hoja de actualizaciones están presentes en el documento global.")

# --- Procesar ejemplares y colapsar filas duplicadas (nueva lógica) ---

# Calcular el número de copias por cada biblionumber ANTES de colapsar
copy_counts = df_filtered_global.groupby('biblionumber_999_c').size()

# Alerta para biblionumber_999_c con más de 3 ejemplares (utilizando copy_counts)
biblio_over_3_copies = copy_counts[copy_counts > 3].index.tolist()

if biblio_over_3_copies:
    print("\n--- ALERTA: Los siguientes biblionumbers tienen más de 3 ejemplares (se mostrarán solo hasta 'Ej. 3') ---")
    for biblio_num in sorted(biblio_over_3_copies):
        print(f"- {biblio_num}: {copy_counts[biblio_num]} ejemplares")
else:
    print("\nNingún biblionumber tiene más de 3 ejemplares.")

# Crear un DataFrame con una sola entrada por biblionumber, tomando la primera fila para el resto de los datos
# (como se solicitó: "solo quedate con la primera que aparece por cada biblionumber")
df_single_entry_global = df_filtered_global.groupby('biblionumber_999_c', as_index=False).first()

# Rellenar las columnas numero_copia_952_t, numero_copia_952_t.1 y numero_copia_952_t.2
# numero_copia_952_t debe ser vacío/NaN
df_single_entry_global['numero_copia_952_t'] = pd.NA

# numero_copia_952_t.1 debe ser 'Ej. 2' si hay 2 o más ejemplares
df_single_entry_global['numero_copia_952_t.1'] = df_single_entry_global['biblionumber_999_c'].map(
    lambda x: 'Ej. 2' if copy_counts.get(x, 0) >= 2 else pd.NA
)
# numero_copia_952_t.2 debe ser 'Ej. 3' si hay 3 o más ejemplares
df_single_entry_global['numero_copia_952_t.2'] = df_single_entry_global['biblionumber_999_c'].map(
    lambda x: 'Ej. 3' if copy_counts.get(x, 0) >= 3 else pd.NA
)

print("\nDataFrame procesado con ejemplares colapsados y columnas 'Ej.' actualizadas (primeras 5 filas):")
display(df_single_entry_global[['biblionumber_999_c', 'numero_copia_952_t', 'numero_copia_952_t.1', 'numero_copia_952_t.2']].head())

# Ahora, la integración debe hacerse con df_single_entry_global
df_updated_final = df_single_entry_global.copy()

# Columnas a excluir del proceso de actualización general desde df_new_updates
# Se excluyen las columnas de ejemplares porque su valor es generado por la lógica anterior.
columns_to_exclude_from_update = [
    'biblionumber_999_c',
    'numero_copia_952_t',
    'numero_copia_952_t.1',
    'numero_copia_952_t.2'
]

# Itera sobre cada fila de la hoja de actualizaciones
for index, row_update in df_new_updates.iterrows():
    biblio_to_update = str(row_update['biblionumber_999_c']).strip() # Ensure consistency

    # Encuentra el índice en df_updated_final que coincide con este biblionumber
    # Ahora solo habrá una fila por biblionumber
    target_index = df_updated_final[df_updated_final['biblionumber_999_c'] == biblio_to_update].index

    if not target_index.empty:
        # Itera sobre las columnas de la fila de actualización, excluyendo las especificadas
        for col_name, value_to_update in row_update.drop(columns_to_exclude_from_update, errors='ignore').items():
            # Solo actualiza si el valor de origen no es NaN y la columna existe en el DataFrame final
            if pd.notna(value_to_update) and col_name in df_updated_final.columns:
                df_updated_final.loc[target_index, col_name] = value_to_update

print("\nDataFrame final después de la integración de datos (primeras 5 filas):")
display(df_updated_final.head())
print(f"Número total de filas en el DataFrame final: {len(df_updated_final)}")

# Asigna el DataFrame procesado final a df_normalized_global para que esté disponible en celdas futuras
df_normalized_global = df_updated_final.copy()

# Opcional: Guardar el resultado final a un archivo CSV local
output_filename_final = 'Registros_Actualizados_Integrados.csv'
df_normalized_global.to_csv(output_filename_final, index=False)
print(f"\nEl archivo '{output_filename_final}' con los registros integrados ha sido creado localmente.")
```

# **5. Pasando de google sheets a mrc**



```python
import pandas as pd
from pymarc import Record, Field, MARCWriter, Subfield

df = df_normalized_global # Usamos df_normalized_global, que es el resultado de la integración

output_filename = "koha_Libros_Actualizados_from_GoogleSheet.mrc"
writer = MARCWriter(open(output_filename, "wb"))

for i, row in df.iterrows():
    record = Record()

    # 020 ISBN
    if 'isbn_020' in row and pd.notna(row['isbn_020']):
        record.add_field(
            Field(
                tag='020',
                indicators=[' ',' '],
                subfields=[Subfield('a', str(row['isbn_020']))]
            )
        )

    # 022 ISSN
    if 'issn_022' in row and pd.notna(row['issn_022']):
        record.add_field(
            Field(
                tag='022',
                indicators=[' ',' '],
                subfields=[Subfield('a', str(row['issn_022']))]
            )
        )

    # 024 DL
    if 'dl_024' in row and pd.notna(row['dl_024']):
        record.add_field(
            Field(
                tag='024',
                indicators=[' ',' '],
                subfields=[Subfield('a', str(row['dl_024']))]
            )
        )

    # 040 Fuente de Catalogación
    subfields_040 = []
    if 'fuente_catalogacion_040' in row and pd.notna(row['fuente_catalogacion_040']):
        subfields_040.append(Subfield('c', str(row['fuente_catalogacion_040'])))

    if subfields_040:
        record.add_field(
            Field(
                tag='040',
                indicators=[' ',' '],
                subfields=subfields_040
            )
        )

    # 041 Idioma de Catalogación
    if 'idioma_catalogacion_041' in row and pd.notna(row['idioma_catalogacion_041']):
        record.add_field(
            Field(
                tag='041',
                indicators=[' ',' '],
                subfields=[Subfield('a', str(row['idioma_catalogacion_041']))]
            )
        )

    # 082 Clasificación Decimal Dewey
    if 'cd_dewey_082' in row and pd.notna(row['cd_dewey_082']):
        record.add_field(
            Field(
                tag='082',
                indicators=['0','0'],
                subfields=[Subfield('a', str(row['cd_dewey_082']))]
            )
        )

    # 100 Autor Personal
    if 'autor_persona_100' in row and pd.notna(row['autor_persona_100']):
        record.add_field(
            Field(
                tag='100',
                indicators=['1',' '],
                subfields=[Subfield('a', str(row['autor_persona_100']))]
            )
        )

    # 110 Autor Institución
    if 'autor_institucion_110' in row and pd.notna(row['autor_institucion_110']):
        record.add_field(
            Field(
                tag='110',
                indicators=['2',' '],
                subfields=[Subfield('a', str(row['autor_institucion_110']))]
            )
        )

    # 111 Autor Congreso
    if 'autor_congreso_111' in row and pd.notna(row['autor_congreso_111']):
        record.add_field(
            Field(
                tag='111',
                indicators=['2',' '],
                subfields=[Subfield('a', str(row['autor_congreso_111']))]
            )
        )

    # 630 Asiento de Materia - Título Uniforme
    if 'materia_titulo_uniforme_630' in row and pd.notna(row['materia_titulo_uniforme_630']):
        record.add_field(
            Field(
                tag='630',
                indicators=['0','0'],
                subfields=[Subfield('a', str(row['materia_titulo_uniforme_630']))]
            )
        )

    # 245 Título
    if 'titulo_245' in row and pd.notna(row['titulo_245']):
        record.add_field(
            Field(
                tag='245',
                indicators=['1','0'],
                subfields=[Subfield('a', str(row['titulo_245']))]
            )
        )

    # 246 Variantes del Título
    if 'varian_del_título_246' in row and pd.notna(row['varian_del_título_246']):
        record.add_field(
            Field(
                tag='246',
                indicators=['3',' '],
                subfields=[Subfield('a', str(row['varian_del_título_246']))]
            )
        )

    # 250 Mención de Edición
    if 'edicion_250' in row and pd.notna(row['edicion_250']):
        record.add_field(
            Field(
                tag='250',
                indicators=[' ',' '],
                subfields=[Subfield('a', str(row['edicion_250']))]
            )
        )

    # 260 Publicación
    subfields_260 = []
    if 'lugar_publicacion_260' in row and pd.notna(row['lugar_publicacion_260']):
        subfields_260.append(Subfield('a', str(row['lugar_publicacion_260'])))
    if 'editorial_260' in row and pd.notna(row['editorial_260']):
        subfields_260.append(Subfield('b', str(row['editorial_260'])))
    if 'anio_260' in row and pd.notna(row['anio_260']):
        # Corrected: convert directly to string, avoiding int() for potential non-numeric values
        subfields_260.append(Subfield('c', str(row['anio_260'])))

    if subfields_260:
        record.add_field(
            Field(
                tag='260',
                indicators=[' ',' '],
                subfields=subfields_260
            )
        )

    # 300 Descripción Física
    if 'descripcion_fisica_300' in row and pd.notna(row['descripcion_fisica_300']):
        record.add_field(
            Field(
                tag='300',
                indicators=[' ',' '],
                subfields=[Subfield('a', str(row['descripcion_fisica_300']))]
            )
        )

    # 490 Mención de Serie 1
    subfields_490_1 = []
    if 'serie_1_490' in row and pd.notna(row['serie_1_490']):
        subfields_490_1.append(Subfield('a', str(row['serie_1_490'])))
    if 'serie_1_490_v' in row and pd.notna(row['serie_1_490_v']):
        subfields_490_1.append(Subfield('v', str(row['serie_1_490_v'])))
    if subfields_490_1:
        record.add_field(
            Field(
                tag='490',
                indicators=['1',' '],
                subfields=subfields_490_1
            )
        )

    # 490 Mención de Serie 2
    subfields_490_2 = []
    if 'serie_2_490' in row and pd.notna(row['serie_2_490']):
        subfields_490_2.append(Subfield('a', str(row['serie_2_490'])))
    if 'serie_2_490_v' in row and pd.notna(row['serie_2_490_v']):
        subfields_490_2.append(Subfield('v', str(row['serie_2_490_v'])))
    if subfields_490_2:
        record.add_field(
            Field(
                tag='490',
                indicators=['1',' '],
                subfields=subfields_490_2
            )
        )

    # 500 Nota General
    if 'nota_general_500' in row and pd.notna(row['nota_general_500']):
        record.add_field(
            Field(
                tag='500',
                indicators=[' ',' '],
                subfields=[Subfield('a', str(row['nota_general_500']))]
            )
        )

    # 505 Formatted Contents Note
    if 'formatted_contents_note_505' in row and pd.notna(row['formatted_contents_note_505']):
        record.add_field(
            Field(
                tag='505',
                indicators=['0',' '],
                subfields=[Subfield('a', str(row['formatted_contents_note_505']))]
            )
        )

    # 520 Resumen
    if 'resumen_520' in row and pd.notna(row['resumen_520']):
        record.add_field(
            Field(
                tag='520',
                indicators=[' ',' '],
                subfields=[Subfield('a', str(row['resumen_520']))]
            )
        )

    # 600 Asiento de Materia - Persona
    for col_suffix in ['1', '2', '3', '4']:
        col_name = f'materia_persona_{col_suffix}_600'
        if col_name in row and pd.notna(row[col_name]):
            record.add_field(
                Field(
                    tag='600',
                    indicators=['1','0'],
                    subfields=[Subfield('a', str(row[col_name]).strip())]
                )
            )

    # 610 Asiento de Materia - Institución
    for col_suffix in ['1', '2', '3', '4']:
        col_name = f'materia_institucion_{col_suffix}_610'
        if col_name in row and pd.notna(row[col_name]):
            record.add_field(
                Field(
                    tag='610',
                    indicators=['2','0'],
                    subfields=[Subfield('a', str(row[col_name]).strip())]
                )
            )

    # 650 Asiento de Materia - Término Temático (Generic)
    for col_suffix in ['1', '2', '3', '4']:
        col_name = f'materia_{col_suffix}'
        if col_name in row and pd.notna(row[col_name]):
            record.add_field(
                Field(
                    tag='650',
                    indicators=[' ','0'],
                    subfields=[Subfield('a', str(row[col_name]).strip())]
                )
            )

    # 651 Asiento de Materia - Término Geográfico
    for col_suffix in ['1', '2']:
        col_name = f'materia_geogra_{col_suffix}'
        if col_name in row and pd.notna(row[col_name]):
            record.add_field(
                Field(
                    tag='651',
                    indicators=[' ','0'],
                    subfields=[Subfield('a', str(row[col_name]).strip())]
                )
            )

    # 655 Término de Indización - Género/Forma
    if 'materia_forma_655' in row and pd.notna(row['materia_forma_655']):
        record.add_field(
            Field(
                tag='655',
                indicators=[' ','7'],
                subfields=[Subfield('a', str(row['materia_forma_655'])), Subfield('2', 'lcgft')]
            )
        )

    # 700 Asiento Secundario - Nombre Personal (coautor_1, coautor_2, coautor_3)
    if 'coautor_1' in row and pd.notna(row['coautor_1']):
        record.add_field(
            Field(
                tag='700',
                indicators=['1',' '],
                subfields=[Subfield('a', str(row['coautor_1'])), Subfield('e', 'coautor')]
            )
        )
    if 'coautor_2' in row and pd.notna(row['coautor_2']):
        record.add_field(
            Field(
                tag='700',
                indicators=['1',' '],
                subfields=[Subfield('a', str(row['coautor_2'])), Subfield('e', 'coautor')]
            )
        )
    if 'coautor_3' in row and pd.notna(row['coautor_3']):
        record.add_field(
            Field(
                tag='700',
                indicators=['1', ' '],
                subfields=[Subfield('a', str(row['coautor_3'])), Subfield('e', 'coautor')]
            )
        )

    # 710 Asiento Secundario - Nombre Corporativo (coau_institución_1, etc.)
    for col_suffix in ['1', '2', '3']:
        col_name = f'coau_institución_{col_suffix}'
        if col_name in row and pd.notna(row[col_name]):
            record.add_field(
                Field(
                    tag='710',
                    indicators=['2',' '],
                    subfields=[Subfield('a', str(row[col_name]).strip()), Subfield('e', 'coautor')]
                )
            )

    # 942 Koha Item Type
    if 'koha_tipo_item_942_c' in row and pd.notna(row['koha_tipo_item_942_c']):
        record.add_field(
            Field(
                tag='942',
                indicators=[' ',' '],
                subfields=[Subfield('c', str(row['koha_tipo_item_942_c']))]
            )
        )

    # 952 Ítems (Múltiples para ejemplares)
    # Capturar subcampos comunes para todos los 952
    common_subfields_for_952 = []
    if 'tipo_item_952_y' in row and pd.notna(row['tipo_item_952_y']):
        common_subfields_for_952.append(Subfield('y', str(row['tipo_item_952_y'])))
    if 'dewey_952_o' in row and pd.notna(row['dewey_952_o']):
        common_subfields_for_952.append(Subfield('o', str(row['dewey_952_o'])))
    if 'cod_barras_952_p' in row and pd.notna(row['cod_barras_952_p']):
        # Ensure cod_barras is an integer if present, otherwise append as string
        try:
            common_subfields_for_952.append(Subfield('p', str(int(row['cod_barras_952_p']))))
        except ValueError:
            common_subfields_for_952.append(Subfield('p', str(row['cod_barras_952_p'])))
    if '952_a' in row and pd.notna(row['952_a']):
        common_subfields_for_952.append(Subfield('a', str(row['952_a'])))
    if '952_b' in row and pd.notna(row['952_b']):
        common_subfields_for_952.append(Subfield('b', str(row['952_b'])))

    # Columnas para los números de copia. No se define default_t_value aquí,
    # el comportamiento de 't' vacío se gestiona a continuación.
    copy_columns_to_process = ['numero_copia_952_t', 'numero_copia_952_t.1', 'numero_copia_952_t.2']

    at_least_one_952_generated = False

    for idx, col_name in enumerate(copy_columns_to_process):
        t_subfield_value = None

        if col_name in row and pd.notna(row[col_name]):
            t_subfield_value = str(row[col_name])

        if t_subfield_value is not None:
            # Si hay un valor explícito para 't', se añade un 952 con los subcampos comunes y el 't'
            current_copy_subfields_952 = list(common_subfields_for_952)
            current_copy_subfields_952.append(Subfield('t', t_subfield_value))
            record.add_field(
                Field(
                    tag='952',
                    indicators=[' ',' '],
                    subfields=current_copy_subfields_952
                )
            )
            at_least_one_952_generated = True
        elif idx == 0 and not at_least_one_952_generated and common_subfields_for_952:
            # Este es el caso del primer ejemplar (col_name 'numero_copia_952_t') cuando está vacío,
            # pero existen otros subcampos comunes (y, o, p). Se añade un 952 sin $t.
            record.add_field(
                Field(
                    tag='952',
                    indicators=[' ',' '],
                    subfields=list(common_subfields_for_952)
                )
            )
            at_least_one_952_generated = True

    # Si no se generó ningún 952 y no hay ningún subcampo común, entonces no se añade nada.
    # Si se generó al menos uno, o si el idx==0 y hay common_subfields_for_952, ya está cubierto.

    # 999 Biblionumber (Koha specific)
    if 'biblionumber_999_c' in row and pd.notna(row['biblionumber_999_c']):
        # Corrected: convert directly to string, avoiding int() for potential non-numeric values
        record.add_field(
            Field(
                tag='999',
                indicators=[' ',' '],
                subfields=[Subfield('c', str(row['biblionumber_999_c']))]
            )
        )

    writer.write(record)

writer.close()

print(f"Archivo MARC '{output_filename}' generado correctamente desde Google Sheet.")
```
